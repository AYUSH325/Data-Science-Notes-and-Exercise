```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('TITLE')
```

```python
#number of rows, second arg is number of columns. 
#third arg is plot number
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented

#Method 1
#OO figures
fig = plt.figure()

#left, bottom, height, width
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

#0.1 = 10%
#Numbers should be between 0 & 1

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.3, 0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
#Method 2 for rows and columns

#tuple unpacking
fig, axes = plt.subplots(nrows=1,ncols=2)

#iterating through all the axes one by one
# for current_ax in axes:
#     current_ax.plot(x,y)

#choosing an axes
axes[0].plot(x,y)
axes[0].set_title('PLOT 1')

axes[1].plot(y,x)
axes[1].set_title('PLOT 2')

#To prevent overlapping of plots
plt.tight_layout()
```

**Figure Size and DPI**

```python
#figsize arg takes width and height
#DPI is dots per inch

fig = plt.figure(figsize=(8,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.set_title('TITLE')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

#loc parameter to set location for legend box
ax.legend(loc=0)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
#save a figure to file
#can specify DPI here toos

fig.savefig('my_picture.png', dpi=200)
```

**PLOT Appearance**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#RGB Hex Code
#linewidth for thickeness of line also can be called as 'lw'
#alpha is transparency of line, eg: alpha=0.5
#ls=linestyle
#marker and marker size for points on graph
ax.plot(x,y,color='#FF8C00', lw=3, ls='-', marker='o',
        markerfacecolor='blue')

#if you want to show between 0 and 1 in x axis and 0 and 2 in y axis
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
