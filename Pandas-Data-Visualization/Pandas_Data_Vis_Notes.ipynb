```python
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
df1.head()
```

```python
df2 = pd.read_csv('df2')
df2.head()
```

```python
df1['A'].hist(bins=30)
```

```python
#Method 1
df1['A'].plot(kind='hist', bins=30)
```

```python
#Method 2
df1['A'].plot.hist()
```

```python
df2.plot.area(alpha=0.4)
```

```python
#categorical 
df2.plot.bar(stacked=True)
```

```python
#For line plots we need x and y axis
df1['A'].plot.line(x=df1.index, y='B', figsize=(12,3), lw=1)
```

```python
#c is color so we show a 3 levels of infomation where we use column C 
#as colour display
df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
#instead of c we can use s - size
df1.plot.scatter(x='A', y='B', s=df1['C']*30)
```

```python
df2.plot.box()
```
